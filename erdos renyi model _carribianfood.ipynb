___Erdos Renyi Model (Caribbean)code___

```python
#importing the networkx library
>>> import networkx as nx

#importing the matplotlib library for plotting the graph
>>> import matplotlib.pyplot as plt

%matplotlib inline
```

```python
n = 44

p = 0.245
```

```python
G = nx.erdos_renyi_graph(n,p)
G.edges()
```

```python
print(nx.info(G))

      
```

```python
fig = plt.figure(figsize=(10, 10), dpi=200)

# Compute the positions of the nodes
pos = nx.spring_layout(G, k=1)

# Draw the nodes and edges
nx.draw(G, pos, with_labels=True, node_size=100, linewidths=10, node_color="#C4D686")

# Remove axis and adjust margins
plt.axis('off')
plt.subplots_adjust(left=0, bottom=0, right=1, top=1)

# Save the plot to a file
plt.savefig("caribbean__ERM_graph.png", dpi=300)
plt.show()
```
